```python
import numpy as np

# Define input
x = np.array([[1, 1]]);

# Output: (1, 2)
print("X's shape: ", x.shape)

# Define result
y = np.array([[0]])

# Output: (1, 1)
print("Y's shape: ", y.shape)

# Define weights
weights = [
    np.array(
      [
        [0.8, 0.2],
        [0.4, 0.9],
        [0.3, 0.5],
      ],
      dtype = np.float32
    ).T,
    np.array(
      [
        [0.3, 0.5, 0.9],
      ],
      dtype = np.float32
    ).T
]
```

```python
def compute_nodes(prevNodes, weights, hasSigmoid = True):
  h = np.dot(prevNodes, weights)
  if hasSigmoid == False: return h
  a = 1 / (1 + np.exp(-h))
  return a

def feed_forward(inputs, outputs, weights):
  a = compute_nodes(inputs, weights[0])
  yHat = compute_nodes(a, weights[1], False)
  mse = np.mean(np.square(yHat - outputs))
  return mse

mse = feed_forward(x, y, weights)
print("MSE: ", mse)
```

```python
from copy import deepcopy

def update_weights(inputs, outputs, weights, learning_rate):
  original_weights = deepcopy(weights)
  temp_weights = deepcopy(weights)
  updated_weights = deepcopy(weights)
  loss = feed_forward(inputs, outputs, original_weights)
  t = 0.0001

  # Iterate through layers
  for i, layer in enumerate(original_weights):
    for j, _ in np.ndenumerate(layer):
      temp_weights = deepcopy(weights)
      temp_weights[i][j] += t
      new_loss = feed_forward(inputs, outputs, temp_weights)
      grad = (new_loss - loss) / t
      updated_weights[i][j] -= grad * learning_rate

  return updated_weights, loss

new_weights, loss = update_weights(x, y, weights, 0.1)

print("New weights: ", new_weights)
print("Loss: ", loss)
```
